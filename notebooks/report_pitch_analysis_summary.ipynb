# ‚öæ MLB Pitch-by-Pitch Velocity Analysis & Scouting Report

## üß≠ Project Overview

This project analyzes MLB pitch-by-pitch data across multiple seasons to evaluate **pitcher performance** using **velocity-based grading**. It produces detailed insights for pitchers, teams, stadiums, and pitch types ‚Äî helping answer: *Who throws the best pitches, where, and for whom?*

---

## üìä Core Components

### 1. **Pitcher Velocity Grading**
- Velocity difference from league average per pitch type
- Graded on a 5-point scale: A (elite) to F (below average)
- Covers **Starters** and **Relievers** separately

### 2. **Pitch Sequences**
- Extracted full pitch sequences per batter
- Saved by season for trend analysis

### 3. **Team-Level Analysis**
- Total Grade A pitch count per team (Starters + Relievers)
- Seasonal trend plots: top improving vs declining staffs
- Combined velocity-based score = bullpen + rotation

### 4. **Stadium Analysis**
- Velocity profile of each stadium
- Top performers by ballpark
- Normalized comparisons to identify bullpen/starter strengths at home

### 5. **Player-Level Scouting**
- Pitchers ranked by velocity diff per pitch type
- League-wide leaderboard for each pitch type
- Red Sox 2025 heatmap of every pitcher and their pitch arsenal
